### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# create a series of data using groupby(), count() method of aggregation, keep the 'ride_id column'

total_ride_by_type = pyber_data_df.groupby(['type']).count()['ride_id']
total_ride_by_type
```

```python
# 2. Get the total drivers for each city type
# create a Series using groupby() method, sum() method of aggregation, keep the 'driver_count'
total_driver_by_type = city_data_df.groupby(['type']).sum()['driver_count']
total_driver_by_type
```

```python
#  3. Get the total amount of fares for each city type
# create a Series using groupby(), sum() method of aggregation, keep the "fare" column.
total_fare_by_type = pyber_data_df.groupby(['type']).sum()['fare']
total_fare_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
# create a Series use groupby(), sum() method of agregation, keep 'fare' column.

avg_fare_by_type = total_fare_by_type / total_ride_by_type
avg_fare_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fare_by_type / total_driver_by_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_ride_summary_df = pd.DataFrame({
    'Total Rides': total_ride_by_type,
    'Total Drivers': total_driver_by_type,
    'Total Fares': total_fare_by_type,
    'Average Fare per Ride': avg_fare_by_type,
    'Average Fare per Driver': avg_fare_per_driver
})

pyber_ride_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_ride_summary_df.index.name = None
pyber_ride_summary_df
```

```python
# Create a pie chart for 'Total Rides'
# Adding pie chart to help visualize the total rides data.

plot = pyber_ride_summary_df['Total Rides'].plot.pie(y='Total Rides', title="Total Rides per Type", legend=False, \
                   autopct='%1.1f%%', explode=(0, 0, 0.1), \
                   shadow=True, startangle=0)

plt.show()
```

```python
#  8. Format the columns.
pyber_ride_summary_df['Total Rides'] = pyber_ride_summary_df['Total Rides'].map('{:,}'.format)
pyber_ride_summary_df['Total Drivers'] = pyber_ride_summary_df['Total Drivers'].map('{:,}'.format)
pyber_ride_summary_df['Total Fares'] = pyber_ride_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_ride_summary_df['Average Fare per Ride'] = pyber_ride_summary_df['Average Fare per Ride'].map('${:.2f}'.format)
pyber_ride_summary_df['Average Fare per Driver'] = pyber_ride_summary_df['Average Fare per Driver'].map('${:.2f}'.format)

pyber_ride_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Read the merged DataFrame
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
# For each date where the indices are the city type and date.
# Use groupby(['type', 'date']), method of aggregation is sum(), and keep 'fare' column
sum_fares_by_type = pyber_data_df.groupby(['type', 'date']).sum()['fare']
sum_fares_by_type
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_by_type_df = sum_fares_by_type.reset_index()
sum_fares_by_type_df
```

```python
# Examine dtype for sum_fares_type_df
sum_fares_by_type_df.dtypes
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# To get the total fares for each type of city by the date. 
sum_fares_by_type_pivot = sum_fares_by_type_df.pivot(index='date', columns='type', values='fare')
sum_fares_by_type_pivot
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_Jan_April = sum_fares_by_type_pivot.loc['2019-01-01':'2019-04-29']
fares_Jan_April.head(10)
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_Jan_April.index = pd.to_datetime(fares_Jan_April.index)
fares_Jan_April
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_Jan_April.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = fares_Jan_April.resample('w').sum()
weekly_fares_df 
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
# From matplotlib import style
# Use the graph style fivethirtyeight.
# Style.use('fivethirtyeight')

ax = weekly_fares_df.plot(figsize= (20,5))

# title
ax.set_title('Total Fare by City Type')

# x_label
ax.set_xlabel('Month')
# y_label
ax.set_ylabel('Fare ($USD)')

# save figure
plt.savefig('analysis/PyBer_fare_summary.png')

plt.show()
```
